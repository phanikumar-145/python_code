```python
amount = 15000
withdraw_limit = 5000
balance = amount
choices = {
    1: "check_balance",
    2: "withdrawl",
    3: "deposit",
    4: "exit"
}
print(choices)
while balance>0:
  user = int(input("enter your choice:"))
  if user == 1:
    print("your balance is:",balance)
  elif user == 2:
    while withdraw_limit<=5000:
      print("your current balance is:",balance)
      withdraw = int(input("enter amount to withdraw:"))
      if balance>=withdraw and withdraw>0 and withdraw<=withdraw_limit:
        balance-=withdraw
        print("your remaining balance",balance)
        break
      else:
        print("check your balance and try again! limit is 5000 only")
  elif user == 3:
    print("your current balance is:",balance)
    deposit_amount = int(input("enter amount to deposit:"))
    if deposit_amount<0:
      print("min deposit rupees 1")
    else:
      balance+=deposit_amount
      print("your balance is",balance)
  else:
    if user == 4:
      print("thank you for using this atm!")
      break
```
