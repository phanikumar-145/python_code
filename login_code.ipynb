```python
attempts = 3
found = False
while attempts>0:
  print("enter your name:")
  name = input()
  if name != "phani":
    print("invalid name")
    attempts = attempts-1
    if attempts==0:
      print("no attempts left try again after 24 hours!")
      break
    else:
      print("you have only ",attempts," attempts left")
  else:
    attempts_pass =3
    while attempts_pass>0:
      print("enter the password")
      password = int(input())
      if password != 6300:
        print("enter correct password:")
        attempts_pass = attempts_pass-1
        print("you have only",attempts_pass," attempts left")
        if attempts_pass==0:
          print("no attempts left try again after 24 hours")
      else:
        print("login successful!")
        found = True
        break
  if found:
    break
  else:
    print("Better luck next time!")
    break
```
