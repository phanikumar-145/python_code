```python
# Write a Python program to calculate the total cab fare for a ride using functions.
# Requirements
# Base fare = ₹50
# Cost per kilometer = ₹12
# Peak hour charge:
# If distance > 10 km → add ₹100
# Night charge:
# If ride is at night (yes / no) → add ₹50
# Use a function to calculate total fare
# Use for loop only if needed (no while)
# Display a detailed fare breakup
```

```python
base_fare = 50
cost_per_kilometer = 12
print('1.yes,2.no')
ride = int(input("enter the ride is night or not:"))
distance = int(input("enter the distance in kilometers:"))
total = base_fare+(cost_per_kilometer*distance)
if ride == 1:
  if distance>10:
    total = total+150
  else:
    total+=50
else:
  if distance>10:
    total+=100
  else:
    total
print("your total fare is:",total)
```
