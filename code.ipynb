```python
# Input:  [0,1,0,3,12]
# Output: [1,3,12,0,0]
n = [0,1,0,3,12]
print("original list",sorted(n))
print("--------------")
m = []
k = []
for i in n:
  if i != 0:
    m.append(i)
  else:
    k.append(i)
# print(m)
# print(k)
m.extend(k)
print("expexted output",m)
```
